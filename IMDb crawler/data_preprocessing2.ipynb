```python
import urllib.request
from bs4 import BeautifulSoup

def getHtml(url):
    try:
        with urllib.request.urlopen(url) as page:
            html = page.read()
            return html
    except:
        return ""
    
def getGenre(html):
    soup = BeautifulSoup(html, 'html.parser', from_encoding='utf-8')
    # print(html)
    year = soup.find('div', 'subtext')
    return year
```

```python
import re
genre_lst = []
start = False
cnt = 0
movie_genre_20M = {}
with open('./ml-latest/movies.csv','rb') as csvfile:
    for row in csvfile:
        if start is False:
            start = True
            continue
        match = re.match("([^\d]*)(.\d*).*", str(row).split(",")[0])
        movieid = match.group(2)
#         genre_lst += str(row).split(',')[-1][:-5].split('|')
#         print(match.group(2))
        genres = str(row).split(',')[-1][:-5].split('|')
#         if int(movieid) <= 135492:
#             continue
        if genres[0] == "(no genres listed)":
            imdb_num = movie_dict[movieid]
#             print("current movieid: ", movieid)
            html = getHtml("http://www.imdb.com/title/tt%s/"%imdb_num)
            genre = getGenre(html)
            g_l = []
            if genre:
                for g in genre.find_all(href=re.compile("genre")):
                    g_l.append(str(g.string))
            movie_genre_20M[movieid] = g_l
        else:
#             print(genres)
            movie_genre_20M[movieid] = genres
        print(movieid, ':', movie_genre_20M[movieid])
# cnt = 0
# for i in genre_lst:
#     if i == '(no genres listed)':
#         cnt += 1
print(cnt)
print(genres)
```
